```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('Products.csv')
data
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
print(data.isnull().sum())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['Price'], kde=True, bins=30, color='blue')
plt.title('Amount Distribution')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
data['Category'].value_counts().plot(kind='bar', color='orange')
plt.title('Product Category Distribution')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='ProductName', y='Price', data=data)
plt.title('Boxplot')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
data.plot(marker='o', color='green')
plt.title('Product sales')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
